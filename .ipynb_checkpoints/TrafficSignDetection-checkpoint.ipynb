```python
import os
import xml.etree.ElementTree as ET
from PIL import Image
import torch
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
import matplotlib.pyplot as plt
```

```python
class TrafficSignDataset(Dataset):
    def __init__(self, root_dir, transform=None):
        self.root_dir = root_dir
        self.transform = transform
        self.image_files = [f for f in os.listdir(root_dir) if f.endswith('.jpg') or f.endswith('.png')]
        print(f"Found {len(self.image_files)} images in the dataset.")
    
    def __len__(self):
        return len(self.image_files)
    
    def __getitem__(self, idx):
        img_name = self.image_files[idx]
        img_path = os.path.join(self.root_dir, img_name)
        image = Image.open(img_path).convert('RGB')
        
        xml_path = os.path.join(self.root_dir, img_name.rsplit('.', 1)[0] + '.xml')
        boxes = []
        labels = []
        
        if os.path.exists(xml_path):
            tree = ET.parse(xml_path)
            root = tree.getroot()
            for obj in root.findall('object'):
                name = obj.find('name').text
                bbox = obj.find('bndbox')
                xmin = float(bbox.find('xmin').text)
                ymin = float(bbox.find('ymin').text)
                xmax = float(bbox.find('xmax').text)
                ymax = float(bbox.find('ymax').text)
                boxes.append([xmin, ymin, xmax, ymax])
                labels.append(name)
        else:
            print(f"Warning: No XML file found for {img_name}")
        
        if self.transform:
            image = self.transform(image)
        
        # Return boxes and labels as lists, not tensors
        return image, {'boxes': boxes, 'labels': labels}

def collate_fn(batch):
    images = [item[0] for item in batch]
    targets = [item[1] for item in batch]
    
    # Stack images
    images = torch.stack(images, 0)
    
    # We don't stack boxes or labels as they may have different lengths
    return images, targets

def test_dataloader(dataloader, device, num_batches=10):
    total_images = 0
    total_objects = 0
    
    for i, (images, targets) in enumerate(dataloader):
        if i >= num_batches:
            break
        
        images = images.to(device)
        
        batch_size = images.shape[0]
        total_images += batch_size
        batch_objects = sum(len(t['boxes']) for t in targets)
        total_objects += batch_objects
        
        print(f"\nBatch {i+1}")
        print(f"Images shape: {images.shape}")
        print(f"Images device: {images.device}")
        print(f"Number of objects in this batch: {batch_objects}")
        
        # Hin th mt nh t batch
        img = images[0].cpu().permute(1, 2, 0).numpy()
        plt.figure(figsize=(10, 10))
        plt.imshow(img)
        plt.axis('off')
        plt.title(f"Sample image from batch {i+1}")
        plt.show()

        # In thng tin v boxes ca nh u tin
        print(f"Boxes for the first image in batch {i+1}:")
        for j, box in enumerate(targets[0]['boxes']):
            print(f"  Box {j+1}: {box}, Label: {targets[0]['labels'][j]}")

    print(f"\nSummary:")
    print(f"Total images processed: {total_images}")
    print(f"Total objects detected: {total_objects}")
    print(f"Average objects per image: {total_objects / total_images:.2f}")

# Thit lp device
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")

# nh ngha transforms
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
])

# To dataset v dataloader
dataset_path = 'C:/Users/khiem/Downloads/BaitapaManh/train'
print(f"Loading dataset from: {dataset_path}")
dataset = TrafficSignDataset(root_dir=dataset_path, transform=transform)
dataloader = DataLoader(dataset, batch_size=4, shuffle=True, num_workers=0, collate_fn=collate_fn)

print(f"Dataset size: {len(dataset)} images")
print(f"Number of batches: {len(dataloader)}")

# Chy hm test
test_dataloader(dataloader, device)

# Kim tra mt vi mu d liu
print("\nChecking a few random samples:")
for i in range(3):
    idx = torch.randint(0, len(dataset), (1,)).item()
    img, target = dataset[idx]
    print(f"\nSample {i+1} (index {idx}):")
    print(f"Image shape: {img.shape}")
    print(f"Number of objects: {len(target['boxes'])}")
    print(f"Labels: {target['labels']}")
```
